```python
% matplotlib inline

import pandas as pd
import pandas_datareader.data as web
from datetime import datetime
import json
```

```python
import dbtools
```

```python
df=web.DataReader("TSLA", 'yahoo', datetime(2016,1,1), datetime(2016, 5, 10))
```

```python
dbtools.df2mongo('stocks', 'prices', df, symbol='TSLA', replace=True)
```

```python
df = dbtools.mongo2df('stocks', 'prices', symbol='TSLA')
```

```python
df[['Close', 'Volume']].plot(subplots=True)
```
